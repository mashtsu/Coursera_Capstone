# Monte Carlo Simulation - Forecasting Revenue

The code below is a Monte Carlo simulation showing 1,000 posible scenarios for revenue based on historical revenue figures. As an assumption I took 170 (million) as the mean revenue with a standard deviation of 20 (million). The goal of this exercise is to use historical data to forecast future revenue.

```python
import numpy as np
import matplotlib.pyplot as plt

print ("Libraries Imported")
```

```python
rev_m = 170   # revenue mean based on historical data
rev_stdev = 20   # standard deviation 
iterations = 1000

print ("Parameters Accepted")
```

```python
rev = np.random.normal(rev_m, rev_stdev,iterations)
rev
```

```python
# the 1000 iterations of revenues can be plotted as below
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
# I calculated Cost of Goods Sold as 60% of revenue

COGS = - (rev*np.random.normal(0.6,0.1))

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean() # mean of COGS
```

```python
COGS.std()  # standard deviation of COGS
```

```python
# Gross Profit is calculated as Revenue less COGS. We had already specified COGS as a negative figure hence why I am adding it below

Gross_Profit = rev + COGS
Gross_Profit

plt.figure(figsize=(15,6))
plt.plot(Gross_Profit)
plt.show()
```

```python
max(Gross_Profit)  # Maximum Gross Profit we can expect
```

```python
min(Gross_Profit)   # Minimum gross profit we can expect
```

```python
Gross_Profit.mean()   # Average gross profit we can expect
```

```python
Gross_Profit.std()   # Standard deviation of gross profit
```

```python
# Bar graph representation of gross profit

plt.figure(figsize=(10,6))
plt.hist(Gross_Profit, bins = [40,50,60,70,80,90,100,110,120]);
plt.ylabel("Number of Occurances")
plt.xlabel("Gross Profit $'millions")
plt.title("Gross Profit Iterations")
plt.show()
```
